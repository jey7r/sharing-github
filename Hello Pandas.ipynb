```python
import pandas as pd
a=pd.DataFrame({
    'roll':[1,2,3,4,5],
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'subject_id':['sub1','sub2','sub4','sub6','sub5']})
b = pd.DataFrame(
    {'address':['Delhi', 'Mumbai', 'Pune', 'Bangalore', 'Chennai'],
     'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
     'Contact':['5234234235','1234284234','1234734234','1236234234','1234254234'],})
print(a)
print(b)
```

```python
pd.concat([a,b])
```

```python
pd.concat([a,b], ignore_index=True) 
#ignore_index=True will ignore the index and create a new index
```

```python
pd.concat([a,b], ignore_index=True, axis=0)
```

```python
pd.concat([a,b], ignore_index=True, axis=1)
```

```python
print(a)
print('_______________________________________')
print(b)
```

```python
pd.merge(a,b, on='Name') #merge on the basis of Name
```

```python
pd.merge(a,b, on='Name', how='inner') #inner join
```

```python
pd.merge(a,b, on='Name', how='outer') #outer join
```

```python
pd.merge(a,b, on='Name', how='left') #left join
```

```python
pd.merge(a,b, on='Name', how='right') #right join
```

```python
pd.merge(a,b, on='Name', how='right', indicator=True) #indicator=True will show the merge type
```

```python
import pandas as pd
a = pd.DataFrame({
    'roll':[1,2,3,4,5,6],
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung', 'Anna'],
    'subject_id':['sub1','sub2','sub4','sub6','sub5', 'sub5']})
b = pd.DataFrame(
    {'address':['Delhi', 'Mumbai', 'Pune', 'Bangalore', 'Chennai', 'Kolkata'],
     'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung','Abhay'],
     'Contact':['5234234235','1234284234','1234734234','1236234234','1234254234', '1234254234'],})
print(a)
print('_______________________________________')
print(b)
```

```python
#Merge multiple dataframes
import pandas as pd
import numpy as np
df1 = pd.DataFrame(np.array([
    ['a', 5, 9],
    ['b', 4, 61],
    ['c', 24, 9]]),
    columns=['name', 'attr11', 'attr12'])
df2 = pd.DataFrame(np.array([
    ['a', 5, 19],
    ['b', 14, 16],
    ['c', 4, 9]]),
    columns=['name', 'attr21', 'attr22'])
df3 = pd.DataFrame(np.array([
    ['a', 15, 49],
    ['b', 4, 36],
    ['c', 14, 9]]),
    columns=['name', 'attr31', 'attr32'])

print(df1,df2, df3)
```

```python
pd.merge(pd.merge(df1,df2,on='name'),df3,on='name')
```

```python
df1.merge(df2,on='name').merge(df3,on='name')
```

groupby, pivot analysis, reshapem melt

```python
df=pd.read_excel("/Users/hopper/Downloads/Class 5B (Pivot Class 6).xlsx", sheet_name='Data')
df
```

```python
# Groupby 'Sales Man' and calculate total price 
# df.groupby(row label)[column].Operation
# Group by 'Sales Man' and sum the 'PRICE' column
total_sales_by_salesman = df.groupby('Sales Man')['PRICE'].sum()
print(total_sales_by_salesman)
```

```python
df.groupby('REGION')['Customer Name'].count()
```

```python
df.groupby('Customer Name')['Sales Man'].sum().head(50)
```

```python
report=df.groupby('CATEGORY')['PRICE'].aggregate(['sum', 'min', 'max', 'mean'])
report.to_excel('/Users/hopper/Downloads/report_Groupby.xlsx')
```

```python
# pivot_table = df.pivot_table(
#     index=  Column to use for the index of the pivot table
#     columns= Column to use for the columns of the pivot table
#     values= Column to aggregate values from
#     aggfunc= Aggregation function to apply (default is 'mean')
#     fill_value= Replace missing values with this value

pivot_table = df.pivot_table(index='REGION', columns='CATEGORY', values='PRICE', aggfunc='sum', fill_value=0)
print(pivot_table)
```

```python
report=pivot_table = df.pivot_table(index='REGION', columns='CATEGORY', values='PRICE', aggfunc='sum', fill_value=0)
report.to_excel('/Users/hopper/Downloads/report_pivot.xlsx')
```

```python
# Reshape the DataFrame as per requirement
# (Example: Reshape the DataFrame to have 'REGION' as index and 'CATEGORY' as columns with 'PRICE' as values)
reshaped_df = df.pivot_table(index='REGION', columns='CATEGORY', values='PRICE', aggfunc='sum', fill_value=0)
print(reshaped_df)
```

```python
# Melt the DataFrame to convert from wide to long format
melted_df = pd.melt(df, id_vars=['S.NO.', 'Customer Name', 'REGION', 'CITY', 'CATEGORY', 'ITEM', 'Date', 'Sales Man'], value_vars=['PRICE'], var_name='Attribute', value_name='Value')
print(melted_df)

```
