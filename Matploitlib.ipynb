```python
import matplotlib.pyplot as plt

labels = ['Python', 'JS', 'C#', 'SQL', 'TypeScript']
sizes = [215, 130, 245, 210, 300]
colors = ['#FFA500', '#800080', '#008000', '#0000FF', '#808080']
explode = (0.1, 0, 0, 0, 0)

plt.title('Languages used in Web Development', y=1.08)  # Adjust y value as needed
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'JS', 'C#', 'SQL', 'TypeScript']
sizes = [215, 130, 245, 210, 300]

# Create a figure with 1 row and 2 columns of subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Pie chart on the first subplot
axs[0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[0].set_title('Pie Chart')

# Bar chart on the second subplot
axs[1].bar(labels, sizes)
axs[1].set_title('Bar Chart')
axs[1].set_xlabel('Languages')
axs[1].set_ylabel('Usage')

# Display the figure with both subplots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Python', 'JS', 'C#', 'SQL', 'TypeScript']
sizes = [215, 130, 245, 210, 300]

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Pie chart on the first subplot
axs[0, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0, 0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[0, 0].set_title('Pie Chart')

# Bar chart on the second subplot
axs[0, 1].bar(labels, sizes)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Languages')
axs[0, 1].set_ylabel('Usage')

# Scatter plot on the third subplot
axs[1, 0].scatter(labels, sizes)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('Languages')
axs[1, 0].set_ylabel('Usage')

# Line plot on the fourth subplot
axs[1, 1].plot(labels, sizes)
axs[1, 1].set_title('Line Plot')
axs[1, 1].set_xlabel('Languages')
axs[1, 1].set_ylabel('Usage')

# Display the figure with all subplots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot on the first subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')

# Bar plot on the second subplot
axs[0, 1].bar(x, y)
axs[0, 1].set_title('Bar Plot')

# Scatter plot on the third subplot
axs[1, 0].scatter(x, y)
axs[1, 0].set_title('Scatter Plot')

# Histogram on the fourth subplot
axs[1, 1].hist(y, bins=20)
axs[1, 1].set_title('Histogram')

# Display the figure with all subplots
plt.tight_layout()
plt.show()
```
